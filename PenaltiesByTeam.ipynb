```python
import requests
import json
import pandas as pd
import plotly.express as px
import numpy
```

```python
statSchedule1516 = pd.read_csv('./gameDataBySeason/NHL20152016.csv')
statSchedule1617 = pd.read_csv('./gameDataBySeason/NHL20162017.csv')
statSchedule1718 = pd.read_csv('./gameDataBySeason/NHL20172018.csv')
statSchedule1819 = pd.read_csv('./gameDataBySeason/NHL20182019.csv')
statSchedule1920 = pd.read_csv('./gameDataBySeason/NHL20192020.csv')
statSchedule2021 = pd.read_csv('./gameDataBySeason/NHL20202021.csv')
```

```python
def initializeEmptyTeamDict():
    teamsData = requests.get('https://statsapi.web.nhl.com/api/v1/teams')
    teamsData = json.loads(teamsData.text)
    teamsData = teamsData["teams"]

    teamsDict = dict()

    for team in teamsData:
        # french accents cause issues with the csv file so changing the name to english
        if team['id'] == 8:
            teamsDict['Montreal Canadians'] = 0
        else:
            teamsDict[team['name']] = 0
            
    return teamsDict
```

```python
def getPIMAndGP1516():
    leaguePIM1516 = initializeEmptyTeamDict()
    leagueGP1516 = initializeEmptyTeamDict()
    
    for index, row in statSchedule1516.iterrows():
        homeTeam = row['homeTeam']
        awayTeam = row['awayTeam']
        leaguePIM1516[homeTeam] += row['homePIM']
        leaguePIM1516[awayTeam] += row['awayPIM']
        leagueGP1516[homeTeam] += 1
        leagueGP1516[awayTeam] += 1
        
    return leaguePIM1516, leagueGP1516
```

```python
# 2015-2016 PIM

leaguePIM1516, leagueGP1516 = getPIMAndGP1516()

del leaguePIM1516['Vegas Golden Knights']
del leaguePIM1516['Seattle Kraken']

teams = leaguePIM1516.keys()
PIM = leaguePIM1516.values()
PIM1516 = {'Teams': teams, 'PIM': PIM}
fig = px.bar(PIM1516, x = 'Teams', y = 'PIM', text = 'PIM')
fig.update_traces(textposition='outside')
fig.update_layout(title_text='PIM of Each Team 2015-2016')
fig.show()

avgPIM = numpy.mean(list(PIM))

mostPenalizedList = {}
for i in range (5):
    mostPenalized = max(leaguePIM1516, key=leaguePIM1516.get)
    mostPenalizedPIM = leaguePIM1516[mostPenalized]
    mostPenalizedList[mostPenalized] = mostPenalizedPIM
    del leaguePIM1516[mostPenalized]

leastPenalizedList = {}
for i in range (5):
    leastPenalized = min(leaguePIM1516, key=leaguePIM1516.get)
    leastPenalizedPIM = leaguePIM1516[leastPenalized]
    leastPenalizedList[leastPenalized] = leastPenalizedPIM
    del leaguePIM1516[leastPenalized]

print('Average season PIM: ' + str(avgPIM))
print('Most penalized teams with cumulative PIM: ' + str(mostPenalizedList))
print('Least penalized teams with cumulative PIM: ' + str(leastPenalizedList))
```

```python
def reportPenaltiesTakenAndDrawn(year):
    if year == '2015-2016':
        seasonPenalties = penalties1516
        leagueGP = getGP1516()
    elif year == '2016-2017':
        seasonPenalties = penalties1617
        leagueGP = getGP1617()
    elif year == '2017-2018':
        seasonPenalties = penalties1718
        leagueGP = getGP1718()
    elif year == '2018-2019':
        seasonPenalties = penalties1819
        leagueGP = getGP1819()
    elif year == '2019-2020':
        seasonPenalties = penalties1920
        leagueGP = getGP1920()
    else:
        seasonPenalties = penalties2021
        leagueGP = getGP2021()
    
    # number of minor penalties taken
    minorsTaken = initializeEmptyTeamDict()
    if year == '2015-2016' or year == '2016-2017':
        del minorsTaken['Vegas Golden Knights']
    del minorsTaken['Seattle Kraken']

    for index, row in seasonPenalties.iterrows():
        if row['penaltySeverity'] == 'Minor':
            minorsTaken[row['offendingTeam']] += 1

    minorsTakenDict = {'Teams': list(minorsTaken.keys()), 'Minor Penalties Taken': list(minorsTaken.values())}
    fig = px.bar(minorsTakenDict, x = 'Teams', y = 'Minor Penalties Taken', text = 'Minor Penalties Taken')
    fig.update_traces(textposition='outside')
    fig.update_layout(title_text='Number of Minor Penalties Taken in the ' + year + ' Season')
    fig.show()

    avgMinorsTaken = numpy.mean(list(minorsTaken.values()))
    mostPenalizedList, leastPenalizedList = getMostAndLeastPenalizedTeams(minorsTaken)

    print('Average number of minor penalties taken: ' + str(avgMinorsTaken))
    print('Most penalized teams by minor penalties taken: ' + str(mostPenalizedList))
    print('Least penalized teams by minor penalties taken: ' + str(leastPenalizedList))

    # number of minor penalties taken per game
    minorsTaken = initializeEmptyTeamDict()
    if year == '2015-2016' or year == '2016-2017':
        del minorsTaken['Vegas Golden Knights']
    del minorsTaken['Seattle Kraken']

    for index, row in seasonPenalties.iterrows():
        if row['penaltySeverity'] == 'Minor':
            minorsTaken[row['offendingTeam']] += 1

    minorsTakenPerGame = initializeEmptyTeamDict()
    if year == '2015-2016' or year == '2016-2017':
        del minorsTakenPerGame['Vegas Golden Knights']
    del minorsTakenPerGame['Seattle Kraken']

    for team in minorsTakenPerGame:
        minorsTakenPerGame[team] = minorsTaken[team] / leagueGP[team]

    minorsTakenPerGameDict = {'Teams': list(minorsTakenPerGame.keys()), 'Minor Penalties Taken Per Game': list(minorsTakenPerGame.values())}
    fig = px.bar(minorsTakenPerGameDict, x = 'Teams', y = 'Minor Penalties Taken Per Game', text = 'Minor Penalties Taken Per Game')
    fig.update_traces(texttemplate='%{text:.2s}', textposition='outside')
    fig.update_layout(title_text='Number of Minor Penalties Taken Per Game in the ' + year + ' Season')
    fig.show()

    avgMinorsTakenPerGame = numpy.mean(list(minorsTakenPerGame.values()))
    mostPenalizedPerGame, leastPenalizedPerGame = getMostAndLeastPenalizedTeams(minorsTakenPerGame)

    print('Average number of minor penalties taken per game: ' + str(avgMinorsTakenPerGame))
    print('Most penalized teams by minor penalties taken per game: ' + str(mostPenalizedPerGame))
    print('Least penalized teams by minor penalties taken per game: ' + str(leastPenalizedPerGame))

    # number of penalties drawn per team
    minorsDrawn = initializeEmptyTeamDict()
    if year == '2015-2016' or year == '2016-2017':
        del minorsDrawn['Vegas Golden Knights']
    del minorsDrawn['Seattle Kraken']

    for index, row in seasonPenalties.iterrows():
        if row['penaltySeverity'] == 'Minor':
            if row['drawnByTeam'] in minorsDrawn.keys():
                minorsDrawn[row['drawnByTeam']] += 1

    minorsDrawnDict = {'Teams': list(minorsDrawn.keys()), 'Minor Penalties Drawn': list(minorsDrawn.values())}
    fig = px.bar(minorsDrawnDict, x = 'Teams', y = 'Minor Penalties Drawn', text = 'Minor Penalties Drawn')
    fig.update_traces(textposition='outside')
    fig.update_layout(title_text='Number of Minor Penalties Drawn in the ' + year + ' Season')
    fig.show()

    avgMinorsDrawn = numpy.mean(list(minorsDrawn.values()))
    mostDrawnList, leastDrawnList = getMostAndLeastPenalizedTeams(minorsDrawn)

    print('Average number of minor penalties drawn: ' + str(avgMinorsDrawn))
    print('Teams with the most minor penalties drawn: ' + str(mostDrawnList))
    print('Teams with the least minor penalties drawn: ' + str(leastDrawnList))

    # number of penalties drawn per game
    minorsDrawnPerGame = initializeEmptyTeamDict()
    if year == '2015-2016' or year == '2016-2017':
        del minorsDrawnPerGame['Vegas Golden Knights']
    del minorsDrawnPerGame['Seattle Kraken']

    for index, row in seasonPenalties.iterrows():
        if row['penaltySeverity'] == 'Minor':
            if row['drawnByTeam'] in minorsDrawnPerGame.keys():
                minorsDrawnPerGame[row['drawnByTeam']] += 1

    for team in minorsDrawnPerGame:
        minorsDrawnPerGame[team] = minorsDrawnPerGame[team] / leagueGP[team]

    minorsDrawnPerGameDict = {'Teams': list(minorsDrawnPerGame.keys()), 'Minor Penalties Drawn Per Game': list(minorsDrawnPerGame.values())}
    fig = px.bar(minorsDrawnPerGameDict, x = 'Teams', y = 'Minor Penalties Drawn Per Game', text = 'Minor Penalties Drawn Per Game')
    fig.update_traces(texttemplate='%{text:.2s}', textposition='outside')
    fig.update_layout(title_text='Number of Minor Penalties Drawn Per Game in the ' + year + ' Season')
    fig.show()

    avgMinorsDrawnPerGame = numpy.mean(list(minorsDrawnPerGame.values()))
    mostDrawnPerGameList, leastDrawnPerGameList = getMostAndLeastPenalizedTeams(minorsDrawnPerGame)

    print('Average number of minor penalties drawn per game: ' + str(avgMinorsDrawnPerGame))
    print('Teams with the most minor penalties drawn per game: ' + str(mostDrawnPerGameList))
    print('Teams with the least minor penalties drawn per game: ' + str(leastDrawnPerGameList))
```
