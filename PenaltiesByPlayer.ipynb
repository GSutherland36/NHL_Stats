```python
import requests
import json
import pandas as pd
import plotly.express as px
import numpy
```

```python
penalties1516 = pd.read_csv('./gameDataBySeason/NHLPenaltyData20152016.csv')
penalties1617 = pd.read_csv('./gameDataBySeason/NHLPenaltyData20162017.csv')
penalties1718 = pd.read_csv('./gameDataBySeason/NHLPenaltyData20172018.csv')
penalties1819 = pd.read_csv('./gameDataBySeason/NHLPenaltyData20182019.csv')
penalties1920 = pd.read_csv('./gameDataBySeason/NHLPenaltyData20192020.csv')
penalties2021 = pd.read_csv('./gameDataBySeason/NHLPenaltyData20202021.csv')
penalties2122 = pd.read_csv('./gameDataBySeason/NHLPenaltyData20212022.csv')
```

```python
def initializePlayerPenaltyDict():
    emptyPlayerDict = dict()
    return emptyPlayerDict
```

```python
# returns a dictionary with every player that took a penalty in the given season and the number of penalties they took

def minorsByPlayer(seasonPenalties):
    playerDict = initializePlayerPenaltyDict()
    
    for index, row in seasonPenalties.iterrows():
        if row['penaltySeverity'] == 'Minor':
            penalizedPlayer = row['penaltyOn']
        
            if penalizedPlayer in playerDict:
                playerDict[penalizedPlayer] += 1
            else:
                playerDict[penalizedPlayer] = 1
    
    return playerDict
```

```python
# returns a dictionary with every player that drew a minor penalty in the given season and the number of minors they drew

def minorsDrawnByPlayer(seasonPenalties):
    playerDict = playerDict = initializePlayerPenaltyDict()
    
    for index, row in seasonPenalties.iterrows():
        if row['penaltySeverity'] == 'Minor':
            penalizedPlayer = row['drewBy']
        
            if penalizedPlayer in playerDict:
                playerDict[penalizedPlayer] += 1
            else:
                playerDict[penalizedPlayer] = 1
    
    return playerDict
```

```python
# takes a dictionary with player IDs as keys and number of minor penalties taken/drawn as values
# returns a dictionary with the 50 max elements of playerDict

def getMostPenalizedPlayers(playerDict):
    mostPenalizedList = {} 
    for i in range (50): 
        mostPenalized = max(playerDict, key=playerDict.get) 
        mostPenalizedPIM = playerDict[mostPenalized] 
        mostPenalizedList[mostPenalized] = mostPenalizedPIM 
        del playerDict[mostPenalized]

    return mostPenalizedList
```

```python
#2015-2016 minors taken

penaltyDict1516 = minorsByPlayer(penalties1516)
mostPenalizedPlayers = getMostPenalizedPlayers(penaltyDict1516)

for player in mostPenalizedPlayers:
    query = 'https://statsapi.web.nhl.com/api/v1/people/' + str(player)
    playerInfo = requests.get(query)
    playerInfo = json.loads(playerInfo.text)
    print(playerInfo['people'][0]['fullName'] + ': ' + str(mostPenalizedPlayers[player]) + ' Minor Penalties in the 2015-2016 season')
```

```python
#2015-2016 minors drawn

penaltiesDrawnDict1516 = minorsDrawnByPlayer(penalties1516)
mostDrawnPlayers = getMostPenalizedPlayers(penaltiesDrawnDict1516)

for player in mostDrawnPlayers:
    query = 'https://statsapi.web.nhl.com/api/v1/people/' + str(player)
    playerInfo = requests.get(query)
    playerInfo = json.loads(playerInfo.text)
    print(playerInfo['people'][0]['fullName'] + ': ' + str(mostDrawnPlayers[player]) + ' minor penalties drawn in the 2015-2016 season')
```
